```python
# Dictionary
my_dictionary = {"name": "Zaira", "age": 20}
print("Original dictionary:", my_dictionary)
```

```python
# Retrieve the value associated with the key "name"
name = my_dictionary["name"]
print("Name from dictionary:", name)
```

```python
# Add a new key-value pair
my_dictionary["city"] = "London"
print("Dictionary after adding a new key-value pair:", my_dictionary)
```

```python
# Set
my_set = {1, 2, 3, 3}
print("Set:", my_set)
```

```python
# Add an element to the set
my_set.add(4)
print("Set after adding element:", my_set)
```

```python
number = 8

if number > 5 and number < 10:
  print(f"{number} is greater than 5 and less than 10")

if number > 10 or number < 2:
  print(f"{number} is greater than 10 or less than 2")
else:
  print(f"{number} is not greater than 10 or less than 2")

if number != 5:
  print(f"{number} is not equal to 5")
```

```python
# Get user input for age
user_input = input("Enter your age: ")

# Convert input to an integer
age = int(user_input)

# Evaluate age using if statements
if age < 0:
    print("Age cannot be negative.")
elif age == 0:
    print("You are a newborn.")
elif age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
elif age < 65:
    print("You are an adult.")
else:
    print("You are a senior.")
```

```python
# Iterate over a range of ages (1 to 10)
print("\nIterating over ages from 1 to 10:")
for i in range(1, 11):
    print(i)
```
